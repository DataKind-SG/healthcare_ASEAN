```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./../Data/interim/disease_MY/weekly-dengue.csv")
```

```python
df.head(20)
```

```python
year_week_aggregation = df.groupby(by=['year','week'])['cases'].sum()
```

```python
df = year_week_aggregation.to_frame()
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
colors = ['r', 'g', 'b', 'm','c']
markers = ['*', 's', 'p', '8','h']
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
for i, year in enumerate(range(2010, 2015)):
    denguef_year_data = df[df['year']==year]
    plt.plot(denguef_year_data.week, denguef_year_data.cases, color = colors[i], marker = markers[i], label=str(year))
    plt.legend()
    plt.xlabel('Week number')
    plt.ylabel('Number of cases')
    plt.title('Dengue Fever weekly trend')
```
